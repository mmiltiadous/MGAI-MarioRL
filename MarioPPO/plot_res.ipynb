```python
import os 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
colors = plt.cm.tab20(range(18))
markers = ['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X', '1', '2', '3']
```

```python
plt.figure(figsize=(12, 8))
i = 0
for subpath in os.listdir('gridsearch'):
    path = os.path.join('gridsearch', subpath)
    csv_file = os.path.join(path, 'progress.csv')
    if not os.path.exists(csv_file):
        print(f"{csv_file} Does not exist")
        continue
    if 'linear' not in subpath:
        continue
    
    df = pd.read_csv(csv_file)
    label = subpath.split('ppo_linear_')[1]
    plt.plot(df['time/total_timesteps'], df['rollout/ep_rew_mean'], label=f"{label}", color=colors[i], marker=markers[i]);
    i += 1


plt.title("Results with linear learning rate schedule")
plt.xlabel("Timesteps")
plt.ylabel('Rollout Mean Reward')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
# plt.tight_layout()

plt.savefig('ppo_lin_lr.png', dpi=300, bbox_inches='tight')
```
