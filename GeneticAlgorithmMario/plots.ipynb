```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into pandas DataFrames
df = pd.read_csv('training_log_mut1.csv')
df2 = pd.read_csv('training_log_mut2.csv')

# Create a 1x3 figure for the subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Plot 1: Fitness over Generations
axs[0].plot(df['Generation'], df['MeanFitness'], label='Mean [0.05,0.15]', linestyle='--', color='blue')
axs[0].plot(df['Generation'], df['BestFitness'], label='Best [0.05,0.15]', linestyle='--', color='green')
axs[0].plot(df['Generation'], df['WorstFitness'], label='Worst [0.05,0.15]', linestyle='--', color='red')
axs[0].plot(df['Generation'], df['MedianFitness'], label='Median [0.05,0.15]', linestyle='--', color='cyan')
axs[0].plot(df['Generation'], df['StdFitness'], label='Std [0.05,0.15]', linestyle='--', color='orange')

axs[0].plot(df2['Generation'], df2['MeanFitness'], label='Mean [0.4,0.5]', linestyle='-', color='blue')
axs[0].plot(df2['Generation'], df2['BestFitness'], label='Best [0.4,0.5]', linestyle='-', color='green')
axs[0].plot(df2['Generation'], df2['WorstFitness'], label='Worst [0.4,0.5]', linestyle='-', color='red')
axs[0].plot(df2['Generation'], df2['MedianFitness'], label='Median [0.4,0.5]', linestyle='-', color='cyan')
axs[0].plot(df2['Generation'], df2['StdFitness'], label='Std [0.4,0.5]', linestyle='-', color='orange')

axs[0].set_xlabel('Generation')
axs[0].set_ylabel('Fitness')
axs[0].set_title('Fitness Metrics over Generations for different Mutation Rates')
axs[0].legend()
axs[0].grid(False)

# Plot 2: Diversity over Generations
axs[1].plot(df['Generation'], df['Diversity'], label='[0.05,0.15]', linestyle='--', color='purple')
axs[1].plot(df2['Generation'], df2['Diversity'], label='[0.4,0.5]', linestyle='-', color='purple')
axs[1].set_xlabel('Generation')
axs[1].set_ylabel('Diversity')
axs[1].set_title('Diversity over Generations for different Mutation Rates')
axs[1].legend()
axs[1].grid(False)

# Plot 3: Time Taken per Generation
axs[2].plot(df['Generation'], df['TimeTaken'], label='[0.05,0.15]', linestyle='--', color='magenta')
axs[2].plot(df2['Generation'], df2['TimeTaken'], label='[0.4,0.5]', linestyle='-', color='magenta')
axs[2].set_xlabel('Generation')
axs[2].set_ylabel('Time Taken (seconds)')
axs[2].set_title('Time Taken per Generation for different Mutation Rates')
axs[2].legend()
axs[2].grid(False)

plt.tight_layout()
plt.show()
```
