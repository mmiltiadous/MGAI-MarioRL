```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_path = 'reports/1717562877_f3120/checkpoints/population_g45_t1717561498.pickle'
population = pickle.load(open(population_path, 'rb'))
```

```python
analytics_manager = population.analytics_manager
best_fitness_per_generation = analytics_manager.best_fitness_per_generation
average_fitness_per_generation = analytics_manager.average_fitness_per_generation
species_counts_per_generation = analytics_manager.species_counts_per_generation
average_neurons_per_generation = analytics_manager.average_neurons_per_generation
average_links_per_generation = analytics_manager.average_links_per_generation
```

```python
def plot_fitness_over_generations():
    generations = range(len(best_fitness_per_generation))
    plt.figure(figsize=(10, 5))
    plt.plot(generations, best_fitness_per_generation, label='Best Fitness')
    plt.plot(generations, average_fitness_per_generation, label='Average Fitness')
    plt.xlabel('Generations')
    plt.ylabel('Fitness')
    plt.title('Fitness Over Generations')
    plt.legend()
    plt.show()
plot_fitness_over_generations()
```

```python
def plot_complexity_over_generations():
    generations = range(len(average_neurons_per_generation))

    plt.figure(figsize=(10, 5))
    plt.plot(generations, average_neurons_per_generation, label='Average Neurons')
    plt.plot(generations, average_links_per_generation, label='Average Links')
    plt.xlabel('Generations')
    plt.ylabel('Count')
    plt.title('Complexity Over Generations')
    plt.legend()
    plt.show()

plot_complexity_over_generations()
```

```python
def plot_species_over_generations():
    generations = range(len(species_counts_per_generation))
    plt.figure(figsize=(10, 5))

    for species_id in species_counts_per_generation[-1].keys():
        species_counts = [gen[species_id] if species_id in gen else 0 for gen in species_counts_per_generation]
        plt.plot(generations, species_counts, label=f'Species {species_id}')
    plt.xlabel('Generations')
    plt.ylabel('Number of Genomes')
    plt.title('Species Population Over Generations')
    plt.legend()
    plt.show()

plot_species_over_generations()
```
